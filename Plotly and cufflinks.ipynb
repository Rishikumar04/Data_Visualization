## Plotly and Cufflinks

* Plotly is a library that allows you to create interactive plots

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())

df1.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[45,56,78]})
                   
df2
```

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

### Scatter

```python
df1.iplot(kind='scatter', x='A', y='B', mode='markers', size=10)
```

### Bar plots

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df1.count().iplot(kind='bar')
```

### boxplot

```python
df1.iplot(kind='box')
```

### 3d Surface

```python
df3 = pd.DataFrame({'X':[1,2,3,4,5],'Y':[10,20,30,40,50],'Z':[5,4,3,2,1]})

df3
```

```python
df3.iplot(kind = 'surface', colorscale='rdylbu')
```

### Spread

```python
df1[['A','B']].iplot(kind='spread')
```

### Histogram

```python
df1['A'].iplot(kind='hist',bins=100)
```

### bubble

```python
df1.iplot(kind='bubble', x='A', y='B', size='C', colorscale='rdylbu')
```

### scatter_matrix()
similar to sns.pairplot()

```python
df1.scatter_matrix()
```
